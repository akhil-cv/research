# Archegos Yahoo Finance Time Series

This notebook downloads the Archegos-related names requested from Yahoo Finance and writes CSV files into `data/`.

It uses fallback ticker candidates for symbols that Yahoo no longer serves under their original names, such as legacy or renamed symbols.

```python
from pathlib import Path

import pandas as pd

from archegos_yahoo_timeseries import DATA_DIR, STOCKS, download_all
```

```python
START = '2020-01-01'
END = '2021-04-30'
INTERVAL = '1d'

results = download_all(start=START, end=END, interval=INTERVAL, output_dir=DATA_DIR)
pd.DataFrame(results)
```
